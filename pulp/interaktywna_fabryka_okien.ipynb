```
from pulp import *
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}

dostepnosc_fabryki1_slider = FloatSlider(min=0,max=8, value=1, description="dostepnosc_fabryki1_slider", style=style)
dostepnosc_fabryki2_slider = FloatSlider(min=0,max=24, value=2, description="dostepnosc_fabryki2_slider", style=style)
dostepnosc_fabryki3_slider = FloatSlider(min=0,max=36, value=5, description="dostepnosc_fabryki3_slider", style=style)

def hurt_produkcja_okien(dostepnosc_fabryki1 = 8.00,
    dostepnosc_fabryki2 = 24.00,
    dostepnosc_fabryki3 = 36.00):
    
    prob = LpProblem("Najlepsza zysk z produkcji HURT",LpMaximize)

    x1 = LpVariable("Ile partii P1", 0)
    x2 = LpVariable("Ile partii P2", 0)

    prob += 3*x1 + 5*x2, "Funkcja kosztu"
    prob += x1 <= dostepnosc_fabryki1, "Dostępność fabryki 1"
    prob += 2*x2 <= dostepnosc_fabryki2, "Dostępność fabryki 2"
    prob += 3*x1 + 2*x2 <= dostepnosc_fabryki3, "Dostępność fabryki 3"

    prob.solve()
    print("Status:", LpStatus[prob.status])

    for v in prob.variables():
      print(v.name, "=", v.varValue)
    #print("Zysk wyniesie {} zł  ".format( value(prob.objective)))

    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print("*"*10+"Shadow Prices"+"*"*10)
    print(pd.DataFrame(shadows))

interact(hurt_produkcja_okien,dostepnosc_fabryki1 = dostepnosc_fabryki1_slider,
    dostepnosc_fabryki2 = dostepnosc_fabryki2_slider,
    dostepnosc_fabryki3 = dostepnosc_fabryki3_slider)
```
